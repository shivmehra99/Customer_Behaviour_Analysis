```python
# we prefer medain over mean because it is not affected by outliers 
# instead of just filling all null values with median with null values will fill by median of review rating based on Category 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# converting columns names into standard way

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'},inplace=True)
```

```python
df.columns
```

```python
# Feature Engineering - creating a new column using old column 
labels = ['Young','Adult','Middle_Age','Senior']
df['age_group'] = pd.qcut(df['age'],q = 4, labels=labels)

# qcut - what it does?
# Basically it divede numberical column in number for group i.e q and assign the labels which we define
```

```python
df[['age','age_group']]
```

```python
# craete a one more column named - purchase frequency days
df['frequency_of_purchases'].unique()
```

```python
# replace categorical value with numerical value using mapping function
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']]
```

```python
# check if discount apply column having same values as promo code use have
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# here both columns have smae values so we can go with one column only

df.drop(['promo_code_used'],axis=1,inplace=True)
```
